```python
file_path = r"C:\Users\Prachi Gohil\Downloads\1_movies_per_genre\1_movies_per_genre\Action.csv"
data = pd.read_csv(file_path)

def convert_run_length(run_length):
    hours = 0
    minutes = 0

    if 'h' in run_length:
        hours = int(run_length.split('h')[0].strip())

    if 'min' in run_length:
        minutes = int(run_length.split('min')[0].split()[-1].strip())

    return hours * 60 + minutes

# Apply the function to the 'run_length' column and replace the column
data['run_length'] = data['run_length'].apply(convert_run_length)

data.head(10) #this func completely replaces run_length with new alues, whereas prev one added the new col 'run_length_minutes'
```

                                                          VISUALISATION

```python
#average movie ratings 
plt.figure(figsize=(12,6))
data.groupby('year')['rating'].mean().plot(kind = 'line',color = 'g' , marker = 'o')
plt.title('AVG MOVIE RATINGS OVER THE YEARS',color='b')
plt.xlabel('year')
plt.ylabel('avg rating')
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(12,6))
data.groupby('year')['num_reviews'].sum().plot(kind='line',marker='o',color='r')
plt.title('TOTAL NO.OF REVIEWS OVER THE YEAR',color='g')
plt.xlabel('years')
plt.ylabel('number of reviews')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('year')['run_length'].mean().plot(kind='line',marker='o',color='b')
plt.title('AVG MOVIE LENGTH(in minutes)OVER THE YEARS',color='r')
plt.xlabel('year')
plt.ylabel('average movie length (minutes)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data.groupby('year').size().plot(kind='bar', color='c')
plt.title('NUMBER OF MOVIES RELEASED EACH YEAR')
plt.xlabel('Year',color='b')
plt.ylabel('Number of Movies',color='b')
plt.xticks(rotation=60)
plt.show()
```

```python
# proportion of movies by genre
import seaborn as sns  #used this for colorss only
genre_counts = data['genres'].value_counts()
#piechart
plt.figure(figsize=(6.5,6.5))
plt.pie(genre_counts,labels=genre_counts.index,autopct='%.1f%%',startangle=20,colors=sns.color_palette("Set3", len(genre_counts)))
plt.title('PROPORTION OF MOVIES BY GENRE',color='r')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

```python
top_movies = data.sort_values(by='rating',ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.barh(top_movies['name'],top_movies['rating'],color='pink')
plt.title('TOP 10 MOVIES BY RATING',color='b')
plt.xlabel('rating',color='r')
plt.ylabel('movie',color='r')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Scatter plot between rating and number of raters
plt.figure(figsize=(10, 6))
plt.scatter(data['num_raters'], data['rating'], alpha=0.7, color='orange')
plt.title('RELATION BETN NO.OF RATERS AND RATINGS',color='b')
plt.xlabel('No.of Raters')
plt.ylabel('Rating')
plt.show()
```

```python
import pandas as pd
# data['release_month'] = pd.to_datetime(data['release_date'],errors='coerce').dt.month
# month_counts = data['release_month'].value_counts().sort_index()
data['release_month'] = pd.to_datetime(data['release_date'], format='mixed', errors='coerce').dt.month
month_counts = data['release_month'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.bar(month_counts.index,month_counts.values,color='teal',alpha=0.8)
plt.title('Number of Movies Released by Month')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.xticks(range(1, 13))
plt.show()
```
